```python
## That's great. But now the columns are multi-indexed, which I find confusing...
# I want to merge the two column headers together. I had to google this, but here's what
# I came up with:

# Store the new column header names in this array
cols = []
for i in range(len(taser_data.columns)):
    top = taser_data.columns.get_level_values(0)[i] # This is what I had to google (get_level_values)
    bottom = taser_data.columns.get_level_values(1)[i]
    
    # Get rid of the empty second-level indexes
    if "Unnamed: " in bottom:
        cols.append(top)
        print top
    else:
        cols.append(top + "_" + bottom)
        print top + "_" + bottom
    
print cols
```
