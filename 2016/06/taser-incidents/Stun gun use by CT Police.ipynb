```python
## Let's make a function to convert these to useful heights in inches

def height_to_inches(h):
    print "Converting height '" + str(h) + "'"
    
    h = str(h)
    h = h.strip()
    
    ''' We will say everything before the single quote is feet
    and everything after is inches, and if it doesn't fit this format
    return null '''
    
    if "'" not in h:
        return None

    quote_i = h.index("'")
    feet = h[:quote_i]
    #print int(feet)
    
    ## IF there is no " denoting inches, we're done
    if "\"" not in h:
        return 12 * int(feet)
    
    inches = h[quote_i+1:len(h) - 1]
    #print inches
    
    return int(feet) * 12 + int(inches)

height_to_inches("5'8\"")
#height_to_inches("6'11")

#height_to_inches("Unknown")
```

```python
# Test out the height conversion function
height_to_inches("5'0\"") == 60
```

```python
height_to_inches("0'0\"") == 0
```

```python
height_to_inches("6'0\"") == 72
```

```python
height_to_inches("Unknown") == None
```

```python
tasers["Height"]["Unnamed: 7_level_1"].apply(lambda x: height_to_inches(x))
```

```python
## Deployment type I and Deployment Type II seem to be the best column to determine 
## if people were warned or tasered. The format is definitely not ideal, having a fixed
## Number of columns for each specific taser deployment, so there can only be a record
## of at maximum two deployments. Not sure what happens if more than two officers
## fire tasers. How would that be reflected in this data set?

#(u'Deployment Type I', 'Unnamed: 10_level_1')
#(u'Deployment Type II', 'Unnamed: 11_level_1')

# Here's what the different codes mean, according to the data dictionary:

# This field indicates the second type of taser deployment if there was more than one 
#(either by multiple officers or different methods by the same officer). 

# The first three values indicate that the subject was not stunned

# U=Un-Holster, 
# L=Laser Sight, 
# W=Warning Arc, 

# The following indicate an actual firing of the taser. 

# C=Cartridge, 
# D=Drive Stun
```

```python
# Secondary firings
tasers["Deployment Type II"]["Unnamed: 11_level_1"].value_counts()
```

```python
## So that's good. Every record has at least a primary deployment type
```

```python
## C + D is the number of people actually fired at
tasers[(tasers["Deployment Type I"]["Unnamed: 10_level_1"] == "C") 
             | (tasers["Deployment Type I"]["Unnamed: 10_level_1"] == "D")]\
       ["Deployment Type I"]["Unnamed: 10_level_1"].count()
```

```python
tasers[(tasers["Deployment Type II"]["Unnamed: 11_level_1"] == "C") 
             | (tasers["Deployment Type II"]["Unnamed: 11_level_1"] == "D")]\
       ["Deployment Type I"]["Unnamed: 10_level_1"].count()
```

```python
tasers_weight_outcomes = pd.to_numeric(tasers["Calculated"]["Height"].to_frame(), errors='coerce')
tasers_weight_outcomes["weight"] = pd.to_numeric(tasers["Weight"]["Unnamed: 8_level_1"], errors='coerce')
tasers_weight_outcomes["deployment_primary"] = tasers["Deployment Type I"]["Unnamed: 10_level_1"]
tasers_weight_outcomes["deployment_secondary"] = tasers["Deployment Type II"]["Unnamed: 11_level_1"]

tasers_weight_outcomes.columns=["height","weight","deployment_primary","deployment_secondary"]
```

```python
tasers_weight_outcomes.groupby("deployment_primary").agg({'height':np.median,'weight':np.median}).reset_index()
```

```python
tasers_weight_outcomes.groupby("deployment_primary").agg({'height':np.mean,'weight':np.mean}).reset_index()
```

```python
# U=Un-Holster, 
# L=Laser Sight, 
# W=Warning Arc, 

# C=Cartridge, 
# D=Drive Stun
```

```python
tasers_weight_outcomes.groupby("deployment_secondary").agg({'height':np.median,'weight':np.median}).reset_index()
```

```python
# What about secondary taser deployments? 

tasers_weight_outcomes.groupby("deployment_secondary").agg({'height':np.mean,'weight':np.mean}).reset_index()
```

```python
# Just a reminder here, the totals for each type of outcome. Far fewer people were tased a second time. 

tasers_weight_outcomes["deployment_primary"].value_counts()
```

```python
tasers_weight_outcomes["deployment_secondary"].value_counts()
```

## Bigger people tasered more

From the above calulations it appears the people who were actually fired upon, either with a cartridge or "drive stun" were a little bit heavier on average, but about the same height. 

A little more than 10 percent of subjects were subject to a secondary taser deployment (such as from a second officer). Those people tended to be heavier than the people who were only fired upon once.
