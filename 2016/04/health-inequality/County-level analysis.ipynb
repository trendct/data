### Life expectancy per county, per quartile

```python
## General function
def le_per_quartile(df, gender):
    for q in range (1, 5):
        col = "le_raceadj_q" + str(q) + "_" + gender
        ret = str(float(df[col]))
        print "Quartile " + str(q) + ": " + ret
        
for c in ct["county_name"]:
    print c + " County"
    df = ct[ct["county_name"] == c]
    le_per_quartile(df,"F")
    print 
```

### Gap between wealthiest and poorest by county


```python
## General function
def le_per_quartile(df, gender):
    #for q in range (1, 5):
    col_q4 = "le_raceadj_q4_" + gender
    col_q1 = "le_raceadj_q1_" + gender

    ret = str(float(df[col_q4]) - float(df[col_q1]))
    print " : " + ret
        
# Women
for c in ct["county_name"]:
    print c + " County"
    df = ct[ct["county_name"] == c]
    le_per_quartile(df,"F")
    print 
    
```

```python
# Men
for c in ct["county_name"]:
    print c + " County"
    df = ct[ct["county_name"] == c]
    le_per_quartile(df,"M")
    print
```
